```python
def add_in_order(p, paciente):
    n = len(p)
    pos = n
    for i in range(n):
        if paciente < p[i]:
            pos = i
            break
    p[pos:pos] = [paciente]
p = [2,3,4,6]
add_in_order(p,5)
print(p)
```
