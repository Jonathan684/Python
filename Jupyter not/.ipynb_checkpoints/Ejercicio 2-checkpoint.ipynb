Desarrollar un programa completo en Python, que a través de un menú de opciones, determine los resultados de las siguientes opciones:

1.) Ingrese el nombre de tres empresas de delivery y la cantidad de paquetes que dichas empresas despacharon, determine cual es la empresa con menor cantidad de envíos. Luego determine el promedio entre todos los envíos, por ultimo determine cuantas empresas tienen una cantidad de paquetes superior al promedio
2.) Ingrese un numero, validar que el mismo sea mayor a cero. Luego ingrese un secuencia de números, la carga finaliza cuando el usuario ingresa un cero, determinar cuantos números de la secuencia al calcular la diferencia con el primer numero ingresado son múltiplos de 2 o 3. Por ultimo determine el porcentaje que esa cantidad calculada respecto de la cantidad total de números leído en la secuencia
3.) Terminar el programa



```python
_author_ = "Algoritmos y Estructuras de Datos"
print("Repaso Parcial 1K05")

menu = "Menu de Opciones\n" \
       "---------------------------------------------------\n" \
       "1 -------------- Empresas de Delivery\n" \
       "2 -------------- Secuencias de numeros\n" \
       "3 -------------- Salir\n" \
       "Ingrese su opcion: "

opcion = 0
while opcion != 3:
    opcion = int(input(menu))
    if opcion == 1:
        nombre_delivery_1 = input("Ingrese el nombre de la primer empresa de delivery: ")
        cantidad_envios_1 = int(input("Ingrese la cantidad de paquetes enviados de la primer empresa de delivery: "))
        nombre_delivery_2 = input("Ingrese el nombre de la primer empresa de delivery: ")
        cantidad_envios_2 = int(input("Ingrese la cantidad de paquetes enviados de la primer empresa de delivery: "))
        nombre_delivery_3 = input("Ingrese el nombre de la primer empresa de delivery: ")
        cantidad_envios_3 = int(input("Ingrese la cantidad de paquetes enviados de la primer empresa de delivery: "))

        if cantidad_envios_1 < cantidad_envios_2 and cantidad_envios_1 < cantidad_envios_3:
            empresa_menor_envios = nombre_delivery_1
        elif cantidad_envios_2 < cantidad_envios_3:
            empresa_menor_envios = nombre_delivery_2
        else:
            empresa_menor_envios = nombre_delivery_3

        promedio = (cantidad_envios_1 + cantidad_envios_2 + cantidad_envios_3) / 3
        contador = 0
        if cantidad_envios_1 > promedio:
            contador += 1

        if cantidad_envios_2 > promedio:
            contador += 1

        if cantidad_envios_3 > promedio:
            contador += 1

        print("La empresa de delivery con menor cantidad de envios fue", empresa_menor_envios)
        print("Hay", contador, "empresas que despacharon mas paquetes que el promedio")

    elif opcion == 2:
        # contadores
        cont_diferencia = cont_total = 0
        numero = 0
        while numero <= 0:
            numero = int(input("Ingrese un numero: "))
            if numero <= 0:
                print("Error!!! El numero ingresado es invalido, debe ser mayor a cero")

        valor = int(input("Ingrese un numero de la secuencia (la carga finaliza con un cero): "))
        while valor != 0:
            cont_total += 1
            diferencia = numero - valor
            if diferencia != 0 and (diferencia % 2 == 0 or diferencia % 3 == 0):
                cont_diferencia += 1

            valor = int(input("Ingrese otro numero de la secuencia (la carga finaliza con un cero): "))

        if cont_diferencia >= 0:
            porcentaje = 0
            if cont_total > 0:
                porcentaje = round(cont_diferencia * 100 / cont_total, 2)

            print("La cantidad de numeros cuya diferencia con", numero,
                  "fueron multiplos de 2 y 3 son", cont_diferencia)
            print("y representan el", porcentaje, "% sobre la cantidad total de numeros de la secuencia")
        else:
            print("Usted no ha ingresado informacion para procesar!!!")

    elif opcion != 3:
        print("Usted ha seleccionado una opcion invalida")
    else:
        print("Programa terminado")
```
