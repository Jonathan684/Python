```python
class Insumo:
    def __init__(self,cod=1,nom='',pre=0.0,cant=0):
        self.codigo = cod
        self.nombre = nom
        self.valor = pre
        self.cantidad = cant
    def __str__(self):
        r = ''
        r += '{:<15}'.format('Codigo: '+str(self.codigo))
        r += '{:<15}'.format('Nombre: '+self.nombre)
        r += '{:<18}'.format('Precio: ' + str(self.valor))
        r += '{:<15}'.format('Cantidad: '+ str(self.cantidad))
        return r
I1 = Insumo(2,"j",2.2,12)
print(I1)
```

```python
def ordernar(v):
    n = len(v)
    for i in range(n - 1):
        for j in range(i+1, n):
            if v[i] < v[j]:
                v[i],v[j] = v[j],v[i]
v = []
v.append(1)
v.append(2)
v.append(3)
v.append(4)
print(v)
ordernar(v)
print(v)
```
