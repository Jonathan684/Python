```python
n = 10
numeros = []
for i in range(1,n+1):
    numeros.append(i)
print("Lista original: ", numeros)
## recorrer de izquierda a derecha y mostrar los numeros pares
#print("Valores pares contenidos en la lista: ", '\t')
print("Valores pares contenidos en la lista: ", end='  ')
for i in range (len(numeros)):
    if numeros[i] % 2 == 0:
        print(numeros[i], end= " ")
```

```python
#Eliminando un numero
print(numeros)
del numeros[2] ## se elimina el numero 3 en la posic
print(numeros)
copia = numeros[:]
print(copia)
cortar = copia[1:2]
print(cortar)
cambiar = copia[:5]
print("cambiar: ",cambiar)
```

```python
def read(v):
    #cargar por teclado de un vector de n componentes
    n = len(v)
    print("Cargue los elementos del vector: ")
    for i in range(n):
        v[i] = int(input('casilla[' + str(i)+']:'))
```

```python
def produc(v, k):
    # multiplica por k al arreglo y lo retorna..
    n = len(v)
    for i in range(n):
        v[i] *= k
```

```python
def validate(inf):
    n = inf
    while n <= inf:
        n = int(input("Cantidad de elemntos (>"+str(inf)+"por favor):"))
        if n<= inf:
            print("Error: se pidio mayor a ", inf,'cargue de nuevo')
    return n
def test():
    ## Cargar cantidad de elementos
    n = validate(0)
    #Crear un arreglo de n elementos (valor inicial 0)
    v = n * [0]
    
    #Cargar el arreglo por teclado..
    read(v)
    ##cargar el nmero k por el cual se multiplicar al vector..
    k = int(input('Ingrese el valor de k: '))
    #multiplicar el vector por k 
    produc(v,k)
    ##mostrar el vector final..
    print('El vector qued as :',v)
if __name__ == '__main__':
    test()
```
